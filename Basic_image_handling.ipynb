# Load packages

```python
import numpy as np
import os
from scipy.misc import imread, imresize
import matplotlib.pyplot as plt
%matplotlib inline
print ("PACKAGES LOADED")
```

# Print Current Folder

```python
cwd = os.getcwd()
print ("Current folder is %s" % (cwd) )

# Useful function
def print_typeshape(img):
    print ("Type is %s" % (type(img)))
    print ("Shape is %s" % (img.shape,))
```

# Loat & Plot

```python
# Load
image = imread(cwd + "/../Colon/CAx20/Image_8607.jpg")
print_typeshape(image)
# Plot
plt.figure(0)
plt.imshow(image)
plt.title("Original Image with imread")
plt.draw()
```

# Resize

### Resize

```python
# Resize
image_small = imresize(image, [120, 160, 3])
print_typeshape(image_small)
# Plot
plt.figure(1)
plt.imshow(image_small)
plt.title("Resized Image")
plt.draw()
```

# Grayscale

### Grayscale

```python
# Grayscale
def rgb2gray(rgb):
    if len(rgb.shape) is 3:
        return np.dot(rgb [..., :3], [0.299, 0.586, 0.114])
    else:
        print ("Current Image if Gray!")
        return rgb

image_small_gray = rgb2gray(image_small)
print_typeshape(image_small)

plt.figure(2)
plt.imshow(image_small_gray, cmap=plt.get_cmap("gray"))
plt.title("[imshow] Gray Image")
plt.colorbar()
plt.draw()
```

# Reshape

```python
# Convert to Vector
image_small_gray_rowvec = np.reshape(image_small_gray, (1,-1));
print_typeshape(image_small_gray_rowvec)

# Convert to Matrix
image_small_gray_mtx = np.reshape(image_small_gray_rowvec, (120, 160));
print_typeshape(image_small_gray_mtx)
```

# Plot Reshaped Image

```python
plt.figure(2)
plt.imshow(image_small_gray_mtx, cmap=plt.get_cmap("gray"))
plt.title("[imshow] Gray Image")
plt.colorbar()
plt.draw()
```

# Load from Folder

```python
cwd = os.getcwd()
path = cwd + "/../Colon/CAx20"
valid_exts = [".jpg",".git",".png",".tag",".jpeg"]
print ("Images in %s are: \n %s" % (path, os.listdir(path)))
```

```python
print ("%d files are in %s" % (len(os.listdir(path)), path))
imgs = []
names = []
for f in os.listdir(path):
    # For all files
    ext = os.path.splitext(f)[1]
    # Check type
    if ext.lower() not in valid_exts:
        continue
    fullpath = os.path.join(path,f)
    # Append to Lists
    imgs.append(imread(fullpath))
    names.append(os.path.splitext(f)[0]+os.path.splitext(f)[1])
print ("%d images loaded" % (len(imgs)))
```

# Check Loaded Images

```python
nimgs = len(imgs)
randidx = np.sort(np.random.randint(nimgs, size=3))
print ("Type of 'imgs': ", type(imgs))
print ("Length of 'imgs': ", len(imgs))
for curr_img, curr_name, i \
    in zip([imgs[j] for j in randidx]
           , [names[j] for j in randidx]
           , range(len(randidx))):
    print ("[%d] Type of 'curr_img' : %s" % (i, type(curr_img)))
    print ("     Name is: %s" % (curr_name))
    print ("     Size of 'curr_img': %s" % (curr_img.shape,))
```

# Plot Images

```python
# Plot Images in 'imgs' list
nimags = len(imgs)
randidx = np.random.randint(nimgs, size=3)
for i in randidx:
    curr_img = imgs[i]
    curr_name = names[i]
    plt.figure(i)
    plt.imshow(curr_img)
    plt.title("[" + str(i) +"]" + curr_name)
    plt.draw()
```
